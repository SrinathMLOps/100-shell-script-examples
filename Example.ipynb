# Comparing different classifiers: Decision Tree, KNN, Naive Bayes, SVM

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stt
from scipy.stats import multivariate_normal as mn
```

```python
from scipy.stats import multivariate_normal
N=500 #number of generated samples
# Covariance values
val = 0
cov = [[1, val], [val, 1]]
# Mean of the distribution
mean = [1, 3]

# Defining bivariate Gaussian distribution
x= multivariate_normal.rvs(mean=mean, cov=cov, size=N)

y=np.ones(N)
#print(y)

# Column bind x and y using np.column_stack
group_1 = np.column_stack((x, y))
#print(group_1)
```

```python
# Covariance values
val1 = 0
cov1 = [[1, val1], [val1, 1]]
# Mean of the distribution
mean1 = [3, 2]

# Generate samples from a bivariate Gaussian distribution
x1 = multivariate_normal.rvs(mean=mean1, cov=cov1, size=N, random_state=42)
#print(x1)

y1=np.zeros(N)
print(y1)

# Column bind x and y using np.column_stack
group_2 = np.column_stack((x1, y1))
#print(group_2)
```

```python
# plot
plt.figure()
plt.scatter(group_1[:,0],group_1[:,1], label='Group 1', s=10, c='b')
plt.scatter(group_2[:,0],group_2[:,1], label='Group 2', s=10, c='r')
plt.title('Simulation from Bivariate Gaussian Distribution')
plt.xlabel('x1')
plt.ylabel('x2')
#plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets as dt
import sklearn.model_selection as ms
import sklearn.tree as tr
import sklearn.neighbors as ne
import sklearn.naive_bayes as nb
import sklearn.linear_model as lm
```

```python
# import dataset
data = np.row_stack((group_1, group_2))
#print(data)
final_x=data[:,0:2]
#print(final_x)
final_y=data[:,2]
#print(final_y)
final_x_tr, final_x_tes, final_y_tr, final_y_tes=ms.train_test_split(final_x,final_y,train_size=0.8, random_state=42)
```

```python
# Decision tree
#Create a Decision Tree classifier with the current max_depth
#from sklearn import tree
DT=tr.DecisionTreeClassifier(max_depth=3)
DT.fit(final_x_tr,final_y_tr)
trACC_DT=DT.score(final_x_tr,final_y_tr)
tesACC_DT=DT.score(final_x_tes,final_y_tes)

print('Train Accuracy for Decision Tree=', trACC_DT)
print('Train Accuracy for Decision Tree=', tesACC_DT)
model = DT.fit(final_x_tr, final_y_tr)
text_representation = tr.export_text(DT)
print(text_representation)

# To Visualize Decision Tree
tr.plot_tree(model)

```

```python
#LR=lm.LogisticRegression()
LR=lm.LogisticRegression(max_iter=200)
LR.fit(final_x_tr,final_y_tr)
trACC_LR=LR.score(final_x_tr,final_y_tr)
tesACC_LR=LR.score(final_x_tes,final_y_tes)
print('Train Accuracy for Logistic Regression=',trACC_LR)
print('Test Accuracy for Logistic Regression=', tesACC_LR)
```

```python
# KNN
KNN=ne.KNeighborsClassifier(n_neighbors=5)
KNN.fit(final_x_tr,final_y_tr)
KNN.fit(final_x_tes,final_y_tes)
trACC_KNN=KNN.score(final_x_tr,final_y_tr)
tesACC_KNN=KNN.score(final_x_tes,final_y_tes)

print('Train Accuracy for KNN=',trACC_KNN)
print('Test Accuracy for KNN=', tesACC_KNN)
```

```python
# NB classifier
NB=nb.GaussianNB()
NB.fit(final_x_tr,final_y_tr)
trACC_NB=NB.score(final_x_tr,final_y_tr)
tesACC_NB=NB.score(final_x_tes,final_y_tes)
print('Train_Accuracy for NB=', trACC_NB)
print('Test_Accuracy for NB=', tesACC_NB)
```

```python
# import module
from tabulate import tabulate

# assign data
myresults = [
    ['', 'DT', 'LR', 'KNN', 'NB'],
    ['Train', trACC_DT, trACC_LR, trACC_KNN, trACC_NB],
    ['Test', tesACC_DT, tesACC_LR, tesACC_KNN, tesACC_NB]
]

# display table
print(tabulate(myresults))
```

```python
# Calculate different performance measures on the test data
#print(final_y_tr)
TP = 0
TN = 0
FN = 0
FP = 0
y_pred=KNN.predict(final_x_tes)
#print(y_pred)
#print(len(final_y_tes))
for i in range(len(final_y_tes)):
    if (final_y_tes[i] == 1) and (y_pred[i] == 1):
        TP += 1
    if (final_y_tes[i] == 1) and (y_pred[i] == 0):
        FN += 1
    if (final_y_tes[i] == 0) and (y_pred[i] == 1):
        FP += 1
    if (final_y_tes[i] == 0) and (y_pred[i] == 0):
        TN += 1
print("TP=", TP, "FN=", FN, "FP=", FP, "TN=", TN)
print (TP / (TP + FP), TP / (TP + FN))
```

```python
#Evaluating of the model
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix

#Calculate accuracy
accuracy = accuracy_score(final_y_tes, y_pred)
print("Accuracy:", accuracy)

# Calculate F1 score
f1 = f1_score(final_y_tes, y_pred)
print("F1 Score:", f1)

# Calculate precision and recall
precision = precision_score(final_y_tes, y_pred)
recall = recall_score(final_y_tes, y_pred)
print("Precision:", precision)
print("Recall:", recall)

# Calculate confusion matrix
conf_matrix = confusion_matrix(final_y_tes, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```
